```python
#29th August 2024, (PSD Heat Map with Log scale) Hourly Average (Daily Plot)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the file path and file name
file_path = '/Users/touqeergill/Desktop/PSD (Heat Map).xlsx'

# Load the data from the Excel file
data = pd.read_excel(file_path, index_col='datetime', parse_dates=True)

# Define the seasons by months for Finland
seasons = {
    'Spring': (3, 4, 5),
    'Summer': (6, 7, 8),
    'Autumn': (9, 10, 11),
    'Winter': (12, 1, 2)
}

# Create a DataFrame for each season by filtering data based on month
seasonal_data = {}
time_index = pd.date_range("00:00", "23:00", freq="H").time  # Uniform time index for hourly intervals
for season, months in seasons.items():
    mask = data.index.month.isin(months)
    seasonal_resampled = data[mask].resample('H').mean()  # Resample data to hourly averages
    seasonal_data[season] = seasonal_resampled.groupby(seasonal_resampled.index.time).mean().reindex(time_index, fill_value=0)

# Set up the figure for plotting
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10), sharex=True, sharey=True)

# Determine the range for y-axis ticks based on your data's range
min_diameter_log = np.floor(np.log10(data.columns.min()))
max_diameter_log = np.ceil(np.log10(data.columns.max()))
ticks = np.logspace(min_diameter_log, max_diameter_log, int(max_diameter_log - min_diameter_log) + 1)

# Plot each season's data using a vibrant colormap (similar to 'jet')
for ax, (season, data) in zip(axes.flatten(), seasonal_data.items()):
    # Setup the extent to map data coordinates to axes coordinates
    extent = [0, len(data), min_diameter_log, max_diameter_log]
    im = ax.imshow(np.log10(data.values).T, aspect='auto', interpolation='none', cmap='jet', origin='lower', extent=extent)
    ax.set_title(f"{season} Season", fontweight='bold')
    ax.set_xlabel('Time of Day', fontweight='bold')
    ax.set_ylabel('Particle diameter', fontweight='bold')
    ax.set_yticks(np.log10(ticks))
    ax.set_yticklabels([f"$\\mathbf{{10^{{{int(np.log10(tick))}}}}}$" for tick in ticks], fontsize=12, weight='bold')
    for label in ax.get_xticklabels():
        label.set_weight('bold')  # Make x-axis labels bold
        label.set_fontsize(12)  # Set font size for x-axis labels

# Adjust layout and add colorbar
plt.subplots_adjust(top=0.8)
cbar_ax = fig.add_axes([0.15, 0.9, 0.7, 0.03])

# Set up the ticks for the colorbar based on the data range
colorbar_ticks = np.log10([10, 100, 1000, 10000, 100000])
colorbar_tick_labels = ['$\\mathbf{10}$', '$\\mathbf{100}$', '$\\mathbf{1000}$', '$\\mathbf{10000}$', '$\\mathbf{100000}$']

cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal', ticks=colorbar_ticks)
cbar.set_label('$\\mathbf{Particle\ concentration\ dN/dlog(Dp)\ [cm^{-3}]}$', labelpad=10, fontweight='bold')
cbar.set_ticklabels(colorbar_tick_labels)
cbar.ax.tick_params(labelsize=10)

# Show plot
plt.show()
```
